```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
import utils
```

```python
dataset=pd.read_csv('accident_data.csv')
```

```python
dataset.isnull().values.any()
```

```python
# taking the important part of the data
df=dataset.iloc[:,2:]
```

```python
df
```

DATA PREPROCESSING

```python
#Checking for null values
df.iloc[:].isnull().values.any()
```

```python
#Transforming  1s and 0s to 'Yes' and 'No' respectively
df=df.iloc[:,:].replace([0,1],['No','Yes'])
```

```python
df
```

```python
# apriori libray requires our dataset to be in the form of list of list. so we transform the panda frame into  alist of lists called strategies using utils.data_prepare
strategies=utils.data_prepare(df)
```

```python
strategies
```

```python
#Using the apriori class to create Associative Rules
Rules=list(apriori(strategies, min_support=0.2,min_confidence=0.2))
```

```python
#Extracting out the rules using utils.extract
associationRules=utils.extract(Rules)
```

```python
#Using panda frame to display each rule and its attributes
rules_df=pd.DataFrame(associationRules,columns=['LHS','RHS','Support','Confidence','Lift'])
```

```python
len(rules_df)
```

```python
rules_df
```

```python
# Extracting the ten top rules based on the Lift
rules_df.nlargest(10, 'Lift')
```

```python
# Extracting the ten top rules based on the Support
rules_df.nlargest(10, 'Support')
```

```python
# Extracting the ten top rules based on the Confidence
rules_df.nlargest(10, 'Confidence')
```
